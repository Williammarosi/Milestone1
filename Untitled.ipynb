```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
letters = ["O", "P", "R", "S", "T", "U", "V", "W", "Z", "A"]
helper = "https://en.wikinews.org/"
link_list = []
for l in letters:
    link_list.append('https://en.wikinews.org/w/index.php?title=Category:Politics_and_conflicts&from={}'.format(l))
```

```python
all_titles = []
all_links = []
all_dates = []
all_content = []
all_sources = []

for url,letter in zip(link_list, letters):

    found = False
    page = url
    
    #while not found:
    response = requests.get(page)
    soup = BeautifulSoup(response.content, "html.parser")
    container = soup.find(id="mw-pages")

    data = container.find_all('li')

    links = [helper + link.get('href') for link in container.find_all('a') if link.get('href')[0:7] == "/wiki/{}".format(letter)]
        
    all_links += links
        
    # Added so it only takes titles with the correct first letter
    titles = [m.text for m in data if m.text[0].upper()==letter]
        
    # Append the titles to the running list
    all_titles += titles
        
    # If the last element on the page is not part of our letter we go to the next

```

```python
for article in all_links[0:10]:
    response = requests.get(article)
    soup = BeautifulSoup(response.content, "html.parser")
    container = soup.find('div', class_ = "mw-parser-output")

    # Date
    dat = [d.text for d in container.find_all("strong", class_ = "published")]
    date = [dat]
    all_dates += date

    # Content
    cont = [c.text for c in container.find_all("p")[1:]]
    content = [cont]
    all_content += content

    # Sources
    sour = [s.text for s in container.find_all("span", class_ = "sourceTemplate")]
    sources = [sour]
    all_sources += sources
```
