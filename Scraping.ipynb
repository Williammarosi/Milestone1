```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
letters = ["O", "P", "R", "S", "T", "U", "V", "W", "Z", "A"]
baseLink = "https://en.wikinews.org/"
link_list = []
for l in letters:
    link_list.append('https://en.wikinews.org/w/index.php?title=Category:Politics_and_conflicts&from={}'.format(l))
```

```python
#empty list to capture all of the links and titles
allLinks = []
allTitles = []


for url,letter in zip(link_list, letters):
    page = url
    
    r = requests.get(page)
    s = BeautifulSoup(r.content, "html.parser")
    c = s.find(id="mw-pages")

    # finds all the tags for links with articles
    allData = c.find_all('li')

    # Puts all of the links to articles in a list and adds it to the list of all links
    links = [baseLink + link.get('href') for link in c.find_all('a') if link.get('href')[0:7] == "/wiki/{}".format(letter)]
    allLinks = allLinks + links

    # Checks if the first letter in the title of the article matches the letter tested and adds it to the list of titles
    titles = [x.text for x in allData if x.text[0].upper()==letter]
    allTitles = allTitles + titles

```

```python
#empty lists for all of the Dates, Content and Sources
allDates = []
allContent = []
allSources = []

# Loop that testes for all links/articles
for article in allLinks:
    r = requests.get(article)
    s = BeautifulSoup(r.content, "html.parser")
    c = s.find('div', class_ = "mw-parser-output")

    #dat = [x.text for x in c.find_all("strong", class_ = "published")]
    date = [x.text for x in c.find_all("strong", class_ = "published")]
    allDates = allDates + date

    #First creates a list of all the content for one article and puts it in the list so it matches the coresponding article
    cont = [x.text for x in c.find_all("p")]
    content = [cont]
    allContent = allContent + content

    #First creates a list of all the sources for one article and puts it in the list so it matches the coresponding article
    sour = [x.text for x in c.find_all("span", class_ = "sourceTemplate")]
    sources = [sour]
    allSources = allSources + sources
```

```python
#Creates a table containing all the obtained data and transposes it
datFra = pd.DataFrame([allTitles, allLinks, allDates, allContent, allSources])
transDatFra = datFra.transpose()

transDatFra
```

```python
#Create titles for the columns
transDatFra.columns = ["Title", "Link", "Date", "Content", "Source"]
transDatFra
```
